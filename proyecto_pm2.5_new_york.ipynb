```python
# 2020
consulta_2020 = "SELECT * FROM `pm2.5_new_york_2020`"
df_2020 = pd.read_sql(consulta_2020, con=engine)
df_2020["Date"] = pd.to_datetime(df_2020["Date"])

# 2021
consulta_2021 = "SELECT * FROM `pm2.5_new_york_2021`"
df_2021 = pd.read_sql(consulta_2021, con=engine)
df_2021["Date"] = pd.to_datetime(df_2021["Date"])

# 2022
consulta_2022 = "SELECT * FROM `pm2.5_new_york_2022`"
df_2022 = pd.read_sql(consulta_2022, con=engine)
df_2022["Date"] = pd.to_datetime(df_2022["Date"])

# 2023
consulta_2023 = "SELECT * FROM `pm2.5_new_york_2023`"
df_2023 = pd.read_sql(consulta_2023, con=engine)
df_2023["Date"] = pd.to_datetime(df_2023["Date"])

# 2024
consulta_2024 = "SELECT * FROM `pm2.5_new_york_2024`"
df_2024 = pd.read_sql(consulta_2024, con=engine)
df_2024["Date"] = pd.to_datetime(df_2024["Date"])
```

## **üìå Evoluci√≥n anual de PM2.5 (2020‚Äì2024)**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Consolidar los a√±os
df_total = pd.concat([df_2020, df_2021, df_2022, df_2023, df_2024], ignore_index=True)

# Agrupar por a√±o
df_total['Year'] = df_total['Date'].dt.year
pm25_yearly = df_total.groupby('Year')['Daily Mean PM2.5 Concentration'].mean()

# Graficar
plt.figure(figsize=(10,5))
pm25_yearly.plot(marker='o', linestyle='-')
plt.title('Evoluci√≥n promedio anual de PM2.5 en Nueva York (2020‚Äì2024)')
plt.xlabel('A√±o')
plt.ylabel('PM2.5 (¬µg/m¬≥)')
plt.grid(True)
plt.show()
```

## **üß™ Distribuci√≥n mensual de PM2.5 (Boxplot 2020‚Äì2024)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Combinar todos los a√±os
df_total = pd.concat([df_2020, df_2021, df_2022, df_2023, df_2024], ignore_index=True)

# Extraer mes
df_total['Month'] = df_total['Date'].dt.month

# Graficar distribuci√≥n mensual
plt.figure(figsize=(12,6))
sns.boxplot(data=df_total, x='Month', y='Daily Mean PM2.5 Concentration', palette='coolwarm')
plt.title('Distribuci√≥n mensual de PM2.5 en Nueva York (2020‚Äì2024)', fontsize=14)
plt.xlabel('Mes')
plt.ylabel('PM2.5 (¬µg/m¬≥)')
plt.grid(True)
plt.xticks(ticks=range(12), labels=['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic'])
plt.tight_layout()
plt.savefig('images/boxplot_pm25_month.png', dpi=300)
plt.show()
```

## **üìà 1. Gr√°fico de l√≠neas con medianas mensuales**
- Este tipo de gr√°fico muestra claramente c√≥mo cambian los valores centrales entre meses:

```python
df_total['Month'] = df_total['Date'].dt.month
monthly_median = df_total.groupby('Month')['Daily Mean PM2.5 Concentration'].median()

monthly_median.plot(marker='o', linestyle='-', figsize=(10,5), title='Mediana mensual de PM2.5 (2020‚Äì2024)')
```

## **üü¶ 2. Gr√°fico de barras con medias o medianas**
- A√∫n m√°s claro para presentaciones o an√°lisis comparativos:

```python
monthly_median.plot(kind='bar', color='steelblue', figsize=(10,5), title='Mediana mensual de PM2.5')
```

## **üí° ¬øC√≥mo saber cu√°ntos son realmente outliers?**
- Podemos calcularlos mes por mes as√≠:

```python
outliers_por_mes = {}

for mes in range(1, 13):
    df_mes = df_total[df_total['Month'] == mes]
    q1 = df_mes['Daily Mean PM2.5 Concentration'].quantile(0.25)
    q3 = df_mes['Daily Mean PM2.5 Concentration'].quantile(0.75)
    iqr = q3 - q1
    limite_superior = q3 + 1.5 * iqr
    outliers = df_mes[df_mes['Daily Mean PM2.5 Concentration'] > limite_superior]
    
    outliers_por_mes[mes] = len(outliers)

# Ver cu√°ntos hay por mes
print(outliers_por_mes)
```

## **üìç Versi√≥n inicial: PM2.5 por d√≠a**

```python
df_total['Year'] = df_total['Date'].dt.year

plt.figure(figsize=(14,6))
sns.scatterplot(data=df_total, x='Date', y='Daily Mean PM2.5 Concentration', hue='Year', palette='Spectral', alpha=0.6)
plt.title('PM2.5 en Nueva York (color por a√±o)')
plt.ylabel('Concentraci√≥n (¬µg/m¬≥)')
plt.xlabel('Fecha')
plt.grid(True)
plt.tight_layout()
plt.savefig('images/scatter_pm25_years.png', dpi=300)
plt.show()
```

## **üéØ 1Ô∏è‚É£ Scatterplot PM2.5 vs AQI con color por a√±o**

## üìä Relaci√≥n entre PM2.5 y AQI (2020‚Äì2024)

Este gr√°fico revela c√≥mo la concentraci√≥n de PM2.5 se relaciona con el AQI, un √≠ndice que traduce los niveles contaminantes en riesgos de salud p√∫blica.

üß† Observaciones clave:
- 2023 presenta los valores m√°s altos en PM2.5 y AQI.
- Existe una correlaci√≥n positiva clara.


```python
import seaborn as sns
import matplotlib.pyplot as plt

df_total['Year'] = df_total['Date'].dt.year

plt.figure(figsize=(12,6))
sns.scatterplot(data=df_total, 
                x='Daily Mean PM2.5 Concentration', 
                y='Daily AQI Value',
                hue='Year',
                palette='Spectral',
                alpha=0.6)

plt.title('Relaci√≥n entre PM2.5 y AQI con color por a√±o (2020‚Äì2024)', fontsize=14)
plt.xlabel('Concentraci√≥n PM2.5 (¬µg/m¬≥)')
plt.ylabel('√çndice AQI')
plt.grid(True)
plt.tight_layout()
plt.savefig('images/scatter_pm25_vs_aqi.png', dpi=300)
plt.show()

```

## **üìà 2Ô∏è‚É£ Doble l√≠nea: evoluci√≥n temporal de PM2.5 y AQI**

```python
import matplotlib.pyplot as plt

# Agrupar por d√≠a para promediar si hay m√°s de una estaci√≥n
df_diario = df_total.groupby('Date')[['Daily Mean PM2.5 Concentration', 'Daily AQI Value']].mean()

plt.figure(figsize=(14,6))
plt.plot(df_diario.index, df_diario['Daily Mean PM2.5 Concentration'], label='PM2.5 (¬µg/m¬≥)', color='firebrick', alpha=0.6)
plt.plot(df_diario.index, df_diario['Daily AQI Value'], label='AQI', color='darkblue', alpha=0.6)
plt.title('Evoluci√≥n diaria de PM2.5 y AQI en Nueva York (2020‚Äì2024)', fontsize=14)
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('images/line_pm25_aqi.png', dpi=300)
plt.show()

```

### üìå Conclusi√≥n del gr√°fico temporal de PM2.5 y AQI

Este gr√°fico muestra c√≥mo la calidad del aire en Nueva York ha fluctuado diariamente entre 2020 y 2024, y c√≥mo el **PM2.5 y el AQI se elevan en paralelo**, revelando una correlaci√≥n estrecha entre la concentraci√≥n de part√≠culas y el riesgo para la salud p√∫blica.
